#Sampling from the participants


```
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 200})
plt.rcParams.update({'font.size': 22})


age = np.arange(0,36,1)

hist_age = np.zeros(len(age))
hist_age[35] = 1
hist_age[28] = 1
hist_age[27] = 2
hist_age[26] = 2
hist_age[25] = 4
hist_age[24] = 7
hist_age[23] = 10
hist_age[22] = 4
hist_age[21] = 4
hist_age[19] = 1

plt.bar(age, hist_age)
plt.xlabel('Alter')
plt.ylabel('Anzahl')
plt.show()
```

```
mean = np.sum(age*hist_age)/np.sum(hist_age)
variance = np.sum(age**2*hist_age)/np.sum(hist_age) - mean**2
error_mean = variance/np.sqrt(np.sum(hist_age))

print('mean     = ', mean)
print('variance = ', variance)
print('error(mean) = ', error_mean)
```

```
plt.bar(age, hist_age/np.sum(hist_age))
plt.xlabel('Alter')
plt.ylabel('Verteilung')
plt.plot(age, 1/np.sqrt(2*np.pi*variance)*np.exp(-(age-mean)**2/2/variance))
plt.show()
```
